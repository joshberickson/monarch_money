```python
import pandas as pd
```

```python
# load the CSV with data in one column
file_path = "connect_institution_selected.csv"
df = pd.read_csv(file_path, header=None, names=["raw_data"])
```

```python
# split the single column into multiple columns
df_split = df['raw_data'].str.split('	', expand=True)

# assign column names
df_split.columns = [
    "EVENT_ID",
    "USER_ID",
    "TIMESTAMP",
    "INSTITUTION_NAME",
    "DATA_PROVIDER",
    "TYPE",
    "SOURCE",
    "CREDENTIAL_ID"
]

# save the fixed data to a new CSV
output_path = "fixed_file_selected.csv"
df_split.to_csv(output_path, index=False)

print(f"File has been fixed and saved to {output_path}")
```
